```python
import pandas as pd
file_path='/content/drive/MyDrive/EVEN.xlsx'
df = pd.read_excel(file_path,engine='openpyxl')
```

```python
df.head()
```

```python
# Q1 What is the average total cost of trips for each destination?

average_cost_per_destination = df.groupby('Destination')['Total_Cost'].mean()
average_cost_per_destination
```

```python
# Q2 What is the distribution of payment methods used by customers?

payment_method_distribution = df['Payment_Method'].value_counts(normalize=True) * 100
payment_method_distribution
```

```python
# Q4 Create a new column that calculates the duration of each trip based on Travel_Dates?

import pandas as pd

def convert_travel_dates(date_str):
    try:
        return pd.to_datetime(date_str)
    except (ValueError, pd.errors.ParserError):
        try:

            return pd.to_datetime(date_str.split(' / ')[0])
        except (ValueError, IndexError):

            return pd.NaT


df['Travel_Dates'] = df['Travel_Dates'].apply(convert_travel_dates)



if 'Return_Date' in df.columns:
    df['Return_Date'] = pd.to_datetime(df['Return_Date'])
    df['Trip_Duration'] = (df['Return_Date'] - df['Travel_Dates']).dt.days
else:
    print("Warning: 'Return_Date' column not found. Assuming all trips last one day.")
    df['Trip_Duration'] = 1

df.head()
```

```python
# Q5 Are there any missing values in important columns such as Travel_Package, Payment_Method, or Travel_Agent?
missing_values = df[['Travel_Package', 'Payment_Method', 'Travel_Agent']].isnull().sum()
print(missing_values)


print("\nRows with missing values in key columns:")
print(df[df[['Travel_Package', 'Payment_Method', 'Travel_Agent']].isnull().any(axis=1)])
```
